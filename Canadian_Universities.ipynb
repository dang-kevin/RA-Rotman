```python
import requests
website = requests.get('https://en.wikipedia.org/wiki/List_of_universities_in_Canada').text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(website,'lxml')
print(soup.prettify())
```

```python
table = soup.find('table',{'class': 'wikitable sortable'})
table
```

```python
links = table.findAll('a')
links
```

```python
info = []
for item in links:
    # Get Universities ('University' or 'Université' -- Use 'Universit') or 'College' or 'École' or 'HEC'
    if item.get('title') != None and ('Universit' in item.get('title') or 'College' in item.get('title') or 'École' in item.get('title') or 'HEC' in item.get('title')):
        info.append(item.get('title')) 
        
print(info)
```

```python
import pandas as pd
df = pd.DataFrame()
df['University'] = info

df
```

```python
#export to excel
from pandas import ExcelWriter

writer = ExcelWriter('Canadian_Universities.xlsx')
df.to_excel(writer,'Sheet1')
writer.save()
```
