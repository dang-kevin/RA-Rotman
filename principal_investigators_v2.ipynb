```python
import pandas as pd
from pybliometrics.scopus import ContentAffiliationRetrieval
#import pybliometrics
#pybliometrics.scopus.utils.create_config()

table3 = pd.read_csv('PI_table3.csv')
aff_ids = list(set(table3['affiliation_id']))
table4 = pd.read_csv('PI_table4.csv')
aff_ids_done = table4['affiliation_id']
aff_ids_missing = list(set(aff_ids)-set(aff_ids_done))
#pd.DataFrame(aff_ids_missing).to_csv('PI_missing_aff_ids.csv',index=False)
#aff_ids_missing = pd.read_csv('PI_missing_aff_ids.csv',header=None).iloc[:,0].tolist()
len(aff_ids_missing)
```

```python
# Weekly quota is 5000.
for i in range(1,len(aff_ids_missing)):
    try:
        aff = ContentAffiliationRetrieval(aff_ids_missing[i])
        table4 = pd.DataFrame([[aff_ids_missing[i], aff.affiliation_name, aff.sort_name, aff.org_type, aff.postal_code, aff.city, \
                              aff.state, aff.country, aff.org_domain, aff.org_URL]], \
                              columns = ['affiliation_id', 'affiliation_name', 'sort_name', 'org_type', 'postal_code',\
                                         'city', 'state', 'country', 'org_domain', 'org_URL'])
        table5 = pd.DataFrame(aff.name_variants)
        table5['affiliation_id'] = aff_ids_missing[i]
        # Append to existing files
        table4.to_csv('PI_table4.csv', mode='a', header=False, index=False)
        table5.to_csv('PI_table5.csv', mode='a', header=False, index=False)
        
    except:
        pass
```
